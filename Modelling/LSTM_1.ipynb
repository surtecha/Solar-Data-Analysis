```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import math
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# 1. Data Preparation
# Load the data from the existing CSV file
df = pd.read_csv('/content/sunspot.csv')

# Convert year and month to datetime
df['date'] = pd.to_datetime(df[['year', 'month']].assign(DAY=1))
df = df.drop(['year', 'month'], axis=1)

# Filter training data to include only records until the end of 2019
train_df = df[df['date'] < '2020-01-01'].copy()
print(f"Training data shape: {train_df.shape}")

# Scale the SSN values
scaler = MinMaxScaler(feature_range=(0, 1))
ssn_scaled = scaler.fit_transform(train_df['ssn'].values.reshape(-1, 1))

# Function to create sequences
def create_sequences(data, window_size=72):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

# Create sequences
X, y = create_sequences(ssn_scaled)
print(f"Sequence data shape: X={X.shape}, y={y.shape}")

# Create sample weights to reduce influence of 24th cycle
# 24th cycle approximately spans from 2008 to 2019
weights = np.ones(len(X))
for i in range(len(X)):
    seq_date = train_df['date'].iloc[i+72]  # Date of the target point
    if '2008-01-01' <= str(seq_date) <= '2019-12-31':
        weights[i] = 0.5  # Lower weight for 24th cycle

# Split data into training and validation sets (80% train, 20% validation)
train_size = int(len(X) * 0.8)
X_train, X_val = X[:train_size], X[train_size:]
y_train, y_val = y[:train_size], y[train_size:]
weights_train = weights[:train_size]

print(f"Training set: X_train={X_train.shape}, y_train={y_train.shape}")
print(f"Validation set: X_val={X_val.shape}, y_val={y_val.shape}")

# 2. Model Architecture
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(72, 1)))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dropout(0.2))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mse')
model.summary()

# 3. Training Process
# Reshape inputs to [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))

# Implement callbacks
checkpoint = ModelCheckpoint('ssn_forecast_model.keras',
                            monitor='val_loss',
                            save_best_only=True,
                            mode='min',
                            verbose=1)

early_stopping = EarlyStopping(monitor='val_loss',
                              patience=10,
                              restore_best_weights=True,
                              verbose=1)

# Train the model
history = model.fit(X_train, y_train,
                    epochs=100,
                    batch_size=32,
                    validation_data=(X_val, y_val),
                    callbacks=[checkpoint, early_stopping],
                    sample_weight=weights_train,
                    verbose=1)

# 4. Evaluation and Visualization
# Plot training metrics
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
plt.show()

# Load the best model
model = load_model('ssn_forecast_model.keras')

# Make predictions on the validation set
y_val_pred = model.predict(X_val)

# Inverse transform the scaled values
y_val_orig = scaler.inverse_transform(y_val.reshape(-1, 1))
y_val_pred_orig = scaler.inverse_transform(y_val_pred)

# Calculate metrics
val_rmse = np.sqrt(mean_squared_error(y_val_orig, y_val_pred_orig))
val_mae = mean_absolute_error(y_val_orig, y_val_pred_orig)
val_r2 = r2_score(y_val_orig, y_val_pred_orig)

print(f"Validation Metrics:")
print(f"RMSE: {val_rmse:.2f}")
print(f"MAE: {val_mae:.2f}")
print(f"R²: {val_r2:.4f}")

# Get dates for validation set visualization
val_dates = train_df['date'].iloc[train_size+72:train_size+72+len(y_val)]

# Plot actual vs. predicted values for validation set
plt.figure(figsize=(14, 7))
plt.plot(val_dates, y_val_orig, label='Actual SSN')
plt.plot(val_dates, y_val_pred_orig, label='Predicted SSN')
plt.title('Validation Set: Actual vs Predicted Sunspot Numbers')
plt.xlabel('Date')
plt.ylabel('Sunspot Number')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. Generate future forecast until 2030
# Get the most recent 72 months of data
recent_data = df[df['date'] <= '2019-12-01'].tail(72)
recent_dates = recent_data['date'].values
last_date = recent_dates[-1]

# Scale the recent data
recent_values = scaler.transform(recent_data['ssn'].values.reshape(-1, 1))
input_seq = recent_values.reshape(1, 72, 1)  # Shape: (1, 72, 1)

# Create date range for forecast
forecast_start = pd.to_datetime(last_date) + pd.DateOffset(months=1)
forecast_end = pd.to_datetime('2030-12-01')
forecast_dates = pd.date_range(start=forecast_start, end=forecast_end, freq='MS')
forecast_length = len(forecast_dates)

# Generate predictions
forecast = []
curr_seq = input_seq.copy()

for _ in range(forecast_length):
    # Get prediction for next month
    next_pred = model.predict(curr_seq)[0, 0]  # Extract the scalar value
    forecast.append(next_pred)

    # Update sequence - correctly handling the dimensions
    curr_seq = np.roll(curr_seq, -1, axis=1)  # Shift values to the left
    curr_seq[0, -1, 0] = next_pred  # Set the last value to the new prediction

# Convert to numpy array and inverse transform
forecast = np.array(forecast).reshape(-1, 1)
forecast = scaler.inverse_transform(forecast)

# 6. Plot results
# Create a figure showing historical and forecasted SSN
plt.figure(figsize=(16, 8))

# Plot historical data from 1996 onwards
historical_data = df[df['date'] >= '1996-01-01']
plt.plot(historical_data['date'], historical_data['ssn'], 'b-', label='Historical SSN')

# Plot forecast
plt.plot(forecast_dates, forecast, 'r-', label='Forecasted SSN (2020-2030)')

# Add vertical line to indicate forecast start
forecast_start_date = pd.to_datetime('2020-01-01')
plt.axvline(x=forecast_start_date, color='k', linestyle='--', label='Forecast Start')

plt.title('Sunspot Number (SSN) Historical Data and Forecast', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Sunspot Number', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# 7. Calculate full dataset performance metrics
# Make predictions on the full dataset
X_full = X.reshape((X.shape[0], X.shape[1], 1))
y_full_pred = model.predict(X_full)

# Inverse transform
y_full_orig = scaler.inverse_transform(y.reshape(-1, 1))
y_full_pred_orig = scaler.inverse_transform(y_full_pred)

# Calculate metrics
full_rmse = np.sqrt(mean_squared_error(y_full_orig, y_full_pred_orig))
full_mae = mean_absolute_error(y_full_orig, y_full_pred_orig)
full_r2 = r2_score(y_full_orig, y_full_pred_orig)

print(f"\nFull Dataset Performance Metrics:")
print(f"RMSE: {full_rmse:.2f}")
print(f"MAE: {full_mae:.2f}")
print(f"R²: {full_r2:.4f}")

# Display a comparison of actual vs predicted values for the full training dataset
full_dates = train_df['date'].iloc[72:72+len(y)]

plt.figure(figsize=(14, 7))
plt.plot(full_dates, y_full_orig, label='Actual SSN')
plt.plot(full_dates, y_full_pred_orig, label='Predicted SSN')
plt.title('Full Dataset: Actual vs Predicted Sunspot Numbers')
plt.xlabel('Date')
plt.ylabel('Sunspot Number')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("Model saved as 'ssn_forecast_model.keras'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input, Bidirectional, Concatenate, Lambda
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import tensorflow as tf
from tensorflow.keras.layers import Flatten
from scipy.optimize import curve_fit
import os
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# 1. Data Preparation
# Load the data from the existing CSV file
df = pd.read_csv('/content/sunspot.csv')

# Convert year and month to datetime
df['date'] = pd.to_datetime(df[['year', 'month']].assign(DAY=1))
df = df.drop(['year', 'month'], axis=1)

# Filter training data to include only records until the end of 2019
train_df = df[df['date'] < '2020-01-01'].copy()
print(f"Training data shape: {train_df.shape}")

# Scale the SSN values
scaler = MinMaxScaler(feature_range=(0, 1))
ssn_scaled = scaler.fit_transform(train_df['ssn'].values.reshape(-1, 1))

# Function to detect solar cycle minima for feature engineering
def detect_solar_minima(values, smoothing_window=13):
    """Detect solar cycle minima in the data for cycle-aware feature engineering"""
    # Apply smoothing to reduce noise
    if len(values) < smoothing_window:
        return []

    # Use convolution for smoothing
    weights = np.ones(smoothing_window) / smoothing_window
    smoothed = np.convolve(values.flatten(), weights, 'valid')

    # Find local minima
    minima_idx = []
    for i in range(2, len(smoothed)-2):
        if (smoothed[i] < smoothed[i-1] and smoothed[i] < smoothed[i-2] and
            smoothed[i] < smoothed[i+1] and smoothed[i] < smoothed[i+2] and
            smoothed[i] < 0.3):  # Threshold in normalized space
            minima_idx.append(i + smoothing_window//2)  # Adjust index for valid convolution

    return minima_idx

# Extract cycle minima for feature engineering
minima_idx = detect_solar_minima(ssn_scaled)
print(f"Detected {len(minima_idx)} solar cycle minima")

# Calculate typical cycle characteristics
if len(minima_idx) >= 2:
    cycle_lengths = np.diff(minima_idx)
    avg_cycle_length = int(np.mean(cycle_lengths))
    print(f"Average cycle length: {avg_cycle_length} months")
else:
    # Default to typical 11-year cycle if detection fails
    avg_cycle_length = 132
    print(f"Using default cycle length: {avg_cycle_length} months")

# Function to create cycle-aware features
def add_cycle_features(data, window_size):
    """Add solar cycle phase and expected amplitude features"""
    # Create sequence data
    sequences = []
    for i in range(len(data) - window_size):
        sequences.append(data[i:i+window_size])

    if not sequences:
        return np.array([])

    sequences = np.array(sequences)

    # For each sequence, find the closest cycle minimum
    cycle_phases = np.zeros((len(sequences), window_size, 1))

    for i in range(len(sequences)):
        # Find the position within the data for this sequence
        seq_start_idx = i
        seq_end_idx = i + window_size

        # Find the closest minimum before or within this sequence
        closest_min_idx = None
        closest_min_distance = float('inf')

        for min_idx in minima_idx:
            # Convert to relative position within the sequence
            rel_pos = min_idx - seq_start_idx

            # If minimum is before sequence, use distance from start
            if rel_pos < 0:
                distance = abs(rel_pos)
            # If minimum is within sequence, use its position
            elif rel_pos < window_size:
                distance = 0
            # If minimum is after sequence, use distance from end
            else:
                distance = rel_pos - window_size

            if distance < closest_min_distance:
                closest_min_distance = distance
                closest_min_idx = min_idx

        # If we found a closest minimum
        if closest_min_idx is not None:
            # Calculate cycle phase for each point in the sequence
            for j in range(window_size):
                # Position relative to closest minimum
                abs_pos = seq_start_idx + j
                months_from_min = abs_pos - closest_min_idx

                # Calculate cycle phase (0 to 1)
                # Normalize by typical cycle length and handle phase wrapping
                cycle_phase = (months_from_min % avg_cycle_length) / avg_cycle_length
                cycle_phases[i, j, 0] = cycle_phase

    # Create expected amplitude feature based on typical cycle shape
    # Use a skewed sine function to model asymmetric solar cycle
    expected_amplitude = np.zeros((len(sequences), window_size, 1))

    for i in range(len(sequences)):
        for j in range(window_size):
            phase = cycle_phases[i, j, 0]
            # Asymmetric cycle: faster rise, slower decay
            if phase < 0.3:  # Rising phase (first 30% of cycle)
                expected_val = np.sin(phase * np.pi / 0.3) * 0.5
            else:  # Declining phase (remaining 70% of cycle)
                expected_val = np.cos((phase - 0.3) * np.pi / 0.7) * 0.5 + 0.5
            expected_amplitude[i, j, 0] = expected_val

    # Add trend feature (direction of change)
    trends = np.zeros((len(sequences), window_size, 1))
    for i in range(len(sequences)):
        for j in range(1, window_size):
            trends[i, j, 0] = sequences[i, j, 0] - sequences[i, j-1, 0]

    # Concatenate all features
    features = np.concatenate([
        sequences,                 # Original SSN values
        cycle_phases,             # Cycle phase (0-1)
        expected_amplitude,       # Expected amplitude based on cycle phase
        trends                    # Short-term trend
    ], axis=2)

    return features

# Function to create sequences with a cycle-aligned window
def create_sequences(data, window_size=132):  # 11 years (132 months) to capture full solar cycle
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

# Create basic sequences
X, y = create_sequences(ssn_scaled)
print(f"Sequence data shape: X={X.shape}, y={y.shape}")

# Add cycle-aware features
X_with_features = add_cycle_features(ssn_scaled, window_size=132)
print(f"Enhanced features shape: {X_with_features.shape}")

# If feature creation failed, fall back to basic sequences
if len(X_with_features) == 0:
    X_with_features = X.reshape(X.shape[0], X.shape[1], 1)
    print("Falling back to basic sequences")

# Create sample weights with cycle-aware approach
weights = np.ones(len(X))
for i in range(len(X)):
    seq_date = train_df['date'].iloc[i+132]  # Date of the target point
    # Lower weight for 24th cycle (anomalous cycle)
    if '2008-01-01' <= str(seq_date) <= '2019-12-31':
        weights[i] = 0.8  # Less aggressive downweighting
    # Higher weight for recent data to prioritize recent patterns
    if '2015-01-01' <= str(seq_date) <= '2019-12-31':
        weights[i] *= 1.2
    # Higher weight for solar minima and maxima points to ensure capturing extremes
    target_val = y[i][0]
    if target_val > 0.7 or target_val < 0.3:
        weights[i] *= 1.2

# Split data into training and validation sets using cycle-aware approach
# Find approximately where cycle 23 starts to use as validation boundary
cycle23_start_idx = train_df[train_df['date'] >= '1996-01-01'].index[0]
# Position in our sequences that corresponds to this boundary
seq_boundary = max(0, cycle23_start_idx - 132)
if seq_boundary >= len(X):
    # Fallback to standard split if boundary is invalid
    train_size = int(len(X) * 0.8)
else:
    train_size = seq_boundary

X_train, X_val = X_with_features[:train_size], X_with_features[train_size:]
y_train, y_val = y[:train_size], y[train_size:]
weights_train = weights[:train_size]

print(f"Training set: X_train={X_train.shape}, y_train={y_train.shape}")
print(f"Validation set: X_val={X_val.shape}, y_val={y_val.shape}")

# 2. Define physics-informed loss function
def physics_informed_loss(y_true, y_pred):
    """
    Physics-informed loss function that penalizes:
    1. General prediction errors (MSE)
    2. Underestimating peaks (important for space weather applications)
    3. Negative predictions (physically impossible)
    4. Excessive short-term volatility (solar activity changes gradually)
    """
    # Base MSE loss
    mse = tf.reduce_mean(tf.square(y_true - y_pred))

    # Higher penalty for underestimating high values (peaks)
    peak_mask = tf.cast(y_true > 0.7, 'float32')  # High values in normalized space
    peak_penalty = tf.reduce_mean(tf.square(tf.maximum(0., y_true - y_pred)) * peak_mask * 2.0)

    # Penalty for negative predictions (physically impossible)
    negative_penalty = tf.reduce_mean(tf.square(tf.maximum(0., -y_pred)) * 10.0)

    # Penalty for overestimating very low values (troughs)
    trough_mask = tf.cast(y_true < 0.3, 'float32')
    trough_penalty = tf.reduce_mean(tf.square(tf.maximum(0., y_pred - y_true)) * trough_mask * 1.5)

    return mse + peak_penalty + negative_penalty + trough_penalty

# 3. Create Hybrid Model with physics-aware components
def create_hybrid_model(input_shape):
    """
    Create a hybrid model combining neural network with physics-based components
    """
    # Neural network path
    nn_input = Input(shape=input_shape)

    # Extract the raw SSN values (first feature channel)
    ssn_values = Lambda(lambda x: x[:, :, 0:1])(nn_input)

    # Extract physics features (cycle phase, expected amplitude, trend)
    physics_features = Lambda(lambda x: x[:, :, 1:])(nn_input)

    # Process SSN values with LSTM
    x = Bidirectional(LSTM(64, return_sequences=True))(ssn_values)
    x = Dropout(0.3)(x)
    x = Bidirectional(LSTM(32))(x)
    nn_features = Dropout(0.3)(x)

    # Process physics features
    # Get the most recent values as they're most relevant for prediction
    recent_physics = Lambda(lambda x: x[:, -24:, :])(physics_features)
    physics_flatten = Flatten()(recent_physics)
    physics_processed = Dense(32, activation='relu')(physics_flatten)
    physics_features = Dropout(0.3)(physics_processed)

    # Combine neural network and physics paths
    combined = Concatenate()([nn_features, physics_features])

    # Output layers with physics constraints
    x = Dense(32, activation='relu')(combined)
    outputs = Dense(1, activation='sigmoid')(x)  # Sigmoid to ensure outputs between 0-1

    model = Model(inputs=nn_input, outputs=outputs)
    model.compile(optimizer='adam', loss=physics_informed_loss)

    return model

# Create the hybrid model
model = create_hybrid_model(X_train.shape[1:])
model.summary()

# 4. Implement Enhanced Training Process
# Create training callbacks
checkpoint = ModelCheckpoint('sunspot_hybrid_model.keras',
                            monitor='val_loss',
                            save_best_only=True,
                            mode='min',
                            verbose=1)

early_stopping = EarlyStopping(monitor='val_loss',
                              patience=20,  # Allow more epochs to find optimal solution
                              restore_best_weights=True,
                              verbose=1)

# Learning rate scheduler to reduce LR when plateauing
lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=10,
    min_lr=1e-5,
    verbose=1
)

# Train the model
history = model.fit(X_train, y_train,
                   epochs=200,  # Increase epochs with early stopping
                   batch_size=32,
                   validation_data=(X_val, y_val),
                   callbacks=[checkpoint, early_stopping, lr_scheduler],
                   sample_weight=weights_train,
                   verbose=1)

# 5. Evaluation and Visualization
# Plot training metrics
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
plt.savefig('model_loss.png')
plt.close()

# Load the best model
try:
    model = tf.keras.models.load_model('sunspot_hybrid_model.keras',
                                      custom_objects={'physics_informed_loss': physics_informed_loss})
except:
    # Use the current model if loading fails
    print("Using current model as loading failed")

# Make predictions on the validation set
y_val_pred = model.predict(X_val)

# Inverse transform the scaled values
y_val_orig = scaler.inverse_transform(y_val.reshape(-1, 1))
y_val_pred_orig = scaler.inverse_transform(y_val_pred)

# Calculate metrics
val_rmse = np.sqrt(mean_squared_error(y_val_orig, y_val_pred_orig))
val_mae = mean_absolute_error(y_val_orig, y_val_pred_orig)
val_r2 = r2_score(y_val_orig, y_val_pred_orig)

print(f"Validation Metrics:")
print(f"RMSE: {val_rmse:.2f}")
print(f"MAE: {val_mae:.2f}")
print(f"R²: {val_r2:.4f}")

# Get dates for validation set visualization
val_dates = train_df['date'].iloc[train_size+132:train_size+132+len(y_val)]

# Plot actual vs. predicted values for validation set
plt.figure(figsize=(14, 7))
plt.plot(val_dates, y_val_orig, label='Actual SSN')
plt.plot(val_dates, y_val_pred_orig, label='Predicted SSN')
plt.title('Validation Set: Actual vs Predicted Sunspot Numbers')
plt.xlabel('Date')
plt.ylabel('Sunspot Number')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('validation_results.png')
plt.close()

# 6. Physics-aware Monte Carlo Dropout for uncertainty estimation
def physics_aware_mc_dropout(model, X, n_samples=100):
    """
    Perform Monte Carlo dropout prediction with physics-based uncertainty adjustments
    """
    predictions = []

    # Use the model multiple times with training=True to keep dropout active
    for _ in range(n_samples):
        pred = model(X, training=True)  # Keep dropout active
        predictions.append(pred.numpy())

    # Calculate statistics
    predictions = np.array(predictions)
    mean_prediction = np.mean(predictions, axis=0)
    std_prediction = np.std(predictions, axis=0)

    # Apply physics-based uncertainty adjustments
    for i in range(len(mean_prediction)):
        # Extract cycle phase from input features (assuming 2nd channel is cycle phase)
        if X.shape[2] > 1:  # If we have the cycle phase feature
            cycle_phase = X[0, -1, 1]  # Use the last timestep's cycle phase

            # Higher uncertainty during solar maxima and minima
            if cycle_phase < 0.2 or (0.3 < cycle_phase < 0.5):
                # Near minimum or maximum - less predictable
                std_prediction[i] *= 1.2

        # Higher uncertainty for extreme values
        if mean_prediction[i] > 0.7:  # High SSN
            std_prediction[i] *= 1.25
        elif mean_prediction[i] < 0.3:  # Low SSN
            std_prediction[i] *= 1.15

    return mean_prediction, std_prediction

# 7. Realistic solar cycle model for post-processing
def skewed_solar_cycle(x, amplitude, frequency, phase, rise_ratio, vertical_shift):
    """
    Skewed sine function to model solar cycle asymmetry with faster rise and slower decline

    Parameters:
    - x: Time values
    - amplitude: Maximum height of the cycle
    - frequency: Controls cycle length
    - phase: Shifts the cycle horizontally
    - rise_ratio: Controls asymmetry (rise vs decay)
    - vertical_shift: Vertical offset
    """
    # Convert time to angle
    angle = frequency * (x - phase)

    # Create skewed waveform by stretching different parts of the cycle
    result = np.zeros_like(x, dtype=float)

    # Rising phase (compressed)
    rising_mask = ((angle % (2*np.pi)) < np.pi)
    result[rising_mask] = np.sin(angle[rising_mask] / rise_ratio)

    # Falling phase (stretched)
    falling_mask = ~rising_mask
    result[falling_mask] = np.sin(angle[falling_mask] * rise_ratio)

    # Scale and shift
    return amplitude * 0.5 * (result + 1) + vertical_shift

# 8. Solar physics-based post-processing
def apply_solar_physics_constraints(forecast_raw, dates):
    """
    Apply solar physics constraints to the raw forecast to ensure physical plausibility
    """
    forecast = forecast_raw.copy()

    # Ensure no negative values
    forecast = np.maximum(forecast, 0)

    # Minimum SSN constraints based on historical solar minima
    forecast = np.maximum(forecast, 5)  # Solar minimum rarely below 5 in modern era

    # Smooth excessively volatile predictions (solar activity changes gradually)
    # Smoothing window increases for longer-term predictions (further in the future)
    smoothed_forecast = forecast.copy()
    for i in range(len(forecast)):
        # Progressive smoothing (more smoothing for later predictions)
        window_size = min(5 + i // 12, 13)  # Increase window size over time, max 13 months
        if window_size % 2 == 0:
            window_size += 1  # Ensure odd window size

        if i >= window_size // 2 and i < len(forecast) - window_size // 2:
            smoothed_forecast[i] = np.mean(forecast[i-window_size//2:i+window_size//2+1])

    # Balance between raw forecast and smoothed version
    # More weight to smoothed version as we go further in the future
    blend_weights = np.minimum(0.3 + 0.5 * np.arange(len(forecast)) / len(forecast), 0.8)
    forecast_adjusted = forecast * (1 - blend_weights) + smoothed_forecast * blend_weights

    # Add cycle length constraint if forecasting across a solar minimum
    # Check if we're forecasting into a new cycle
    forecast_dates = pd.to_datetime(dates)
    forecast_years = forecast_dates.year

    # Detect potential solar minimum in the forecast
    # Look for local minima in the smoothed forecast
    potential_minima = []
    for i in range(2, len(smoothed_forecast)-2):
        if (smoothed_forecast[i] < smoothed_forecast[i-1] and
            smoothed_forecast[i] < smoothed_forecast[i-2] and
            smoothed_forecast[i] < smoothed_forecast[i+1] and
            smoothed_forecast[i] < smoothed_forecast[i+2] and
            smoothed_forecast[i] < 20):  # Threshold for solar minimum
            potential_minima.append(i)

    # If we detect a minimum, ensure we have realistic rise in the next cycle
    if potential_minima:
        for min_idx in potential_minima:
            # Only process if we have enough data points after the minimum
            if min_idx < len(forecast_adjusted) - 24:
                # Apply rising phase correction (typically ~4 years from min to max)
                for j in range(min_idx + 1, min(len(forecast_adjusted), min_idx + 48)):
                    months_from_min = j - min_idx

                    # Calculate expected rise based on typical solar cycle
                    # Using 48 months (4 years) as typical rise time
                    if months_from_min < 48:
                        rise_ratio = months_from_min / 48
                        expected_rise = rise_ratio * 100  # Typical rise from min to max (~100 SSN)

                        # Blend with original forecast (more weight to physics model in early rise)
                        blend_factor = max(0.7 - months_from_min / 48, 0)
                        min_val = forecast_adjusted[min_idx]
                        physics_val = min_val + expected_rise

                        # Ensure we don't decrease during the initial rising phase
                        forecast_adjusted[j] = max(
                            forecast_adjusted[j],
                            forecast_adjusted[j-1] * 0.98,  # Allow small decreases (2% max)
                            forecast_adjusted[min_idx] + (months_from_min/6)  # Minimum expected rise
                        )

    return forecast_adjusted

# 9. Generate future forecast until 2030 with physics-based uncertainty
# Get the most recent data for forecast starting point
recent_data = df[df['date'] <= '2019-12-31'].tail(132)
recent_dates = recent_data['date'].values
last_date = recent_dates[-1]

# Prepare input features
recent_values = scaler.transform(recent_data['ssn'].values.reshape(-1, 1))
recent_features = add_cycle_features(recent_values, window_size=132)

# If feature creation failed, use basic input
if len(recent_features) == 0:
    input_seq = recent_values.reshape(1, 132, 1)
else:
    input_seq = recent_features[-1:].reshape(1, 132, -1)

# Create date range for forecast
forecast_start = pd.to_datetime(last_date) + pd.DateOffset(months=1)
forecast_end = pd.to_datetime('2030-12-31')
forecast_dates = pd.date_range(start=forecast_start, end=forecast_end, freq='MS')
forecast_length = len(forecast_dates)

# Generate predictions with physics-aware uncertainty
forecast = []
forecast_lower = []
forecast_upper = []
uncertainty = []
curr_seq = input_seq.copy()

for _ in range(forecast_length):
    try:
        # Get prediction and uncertainty for next month
        mean_pred, std_pred = physics_aware_mc_dropout(model, curr_seq, n_samples=50)
        next_pred = mean_pred[0, 0]  # Extract the scalar value
        next_std = std_pred[0, 0]  # Extract the uncertainty
    except:
        # Fallback to standard prediction if MC dropout fails
        next_pred = model.predict(curr_seq)[0][0]
        next_std = 0.1  # Default uncertainty

    forecast.append(next_pred)
    uncertainty.append(next_std)

    # Calculate prediction intervals (95% confidence)
    forecast_lower.append(max(0, next_pred - 1.96 * next_std))  # Prevent negative values
    forecast_upper.append(next_pred + 1.96 * next_std)

    # Update sequence for next prediction - correctly handling the dimensions
    if curr_seq.shape[2] > 1:  # If using enhanced features
        # Create a new row with updated features
        # This is a simplification - in production you'd compute actual feature values
        new_row = np.zeros((1, 1, curr_seq.shape[2]))
        new_row[0, 0, 0] = next_pred  # Set SSN value

        # Basic update of cycle phase (increment by 1/132)
        if curr_seq.shape[2] > 1:
            new_row[0, 0, 1] = (curr_seq[0, -1, 1] + 1/avg_cycle_length) % 1

        # Update expected amplitude based on new phase
        phase = new_row[0, 0, 1]
        if phase < 0.3:  # Rising phase
            expected_val = np.sin(phase * np.pi / 0.3) * 0.5
        else:  # Declining phase
            expected_val = np.cos((phase - 0.3) * np.pi / 0.7) * 0.5 + 0.5
        if curr_seq.shape[2] > 2:
            new_row[0, 0, 2] = expected_val

        # Update trend
        if curr_seq.shape[2] > 3:
            new_row[0, 0, 3] = next_pred - curr_seq[0, -1, 0]

        # Shift values and add new prediction
        curr_seq = np.concatenate([curr_seq[:, 1:, :], new_row], axis=1)
    else:
        # Simple update for basic sequence
        curr_seq = np.roll(curr_seq, -1, axis=1)
        curr_seq[0, -1, 0] = next_pred

# Convert to numpy arrays
forecast = np.array(forecast).reshape(-1, 1)
forecast_lower = np.array(forecast_lower).reshape(-1, 1)
forecast_upper = np.array(forecast_upper).reshape(-1, 1)
uncertainty = np.array(uncertainty)

# Inverse transform to get actual SSN values
forecast_raw = scaler.inverse_transform(forecast)
forecast_lower = scaler.inverse_transform(forecast_lower)
forecast_upper = scaler.inverse_transform(forecast_upper)

# Apply solar physics constraints
forecast_final = apply_solar_physics_constraints(forecast_raw[:, 0], forecast_dates)

# Also apply constraints to confidence bounds
forecast_lower_final = apply_solar_physics_constraints(forecast_lower[:, 0], forecast_dates)
forecast_upper_final = apply_solar_physics_constraints(forecast_upper[:, 0], forecast_dates)

# Ensure lower bound doesn't exceed forecast and upper bound exceeds forecast
forecast_lower_final = np.minimum(forecast_lower_final, forecast_final)
forecast_upper_final = np.maximum(forecast_upper_final, forecast_final)

# 10. Plot results with prediction intervals
plt.figure(figsize=(16, 8))

# Plot historical data from 1996 onwards
historical_data = df[df['date'] >= '1996-01-01']
plt.plot(historical_data['date'], historical_data['ssn'], 'b-', label='Historical SSN')

# Plot forecast with prediction intervals
plt.plot(forecast_dates, forecast_final, 'r-', label='Forecasted SSN (2020-2030)')
plt.fill_between(forecast_dates,
                forecast_lower_final,
                forecast_upper_final,
                color='red', alpha=0.2,
                label='95% Prediction Interval')

# Add vertical line to indicate forecast start
forecast_start_date = pd.to_datetime('2020-01-01')
plt.axvline(x=forecast_start_date, color='k', linestyle='--', label='Forecast Start')

plt.title('Sunspot Number (SSN) Historical Data and Forecast', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Sunspot Number', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig('ssn_forecast_2030.png')
plt.show()

# 11. Calculate full dataset performance metrics
# Create features for full dataset evaluation
X_full = X_with_features
y_full_pred = model.predict(X_full)

# Inverse transform
y_full_orig = scaler.inverse_transform(y.reshape(-1, 1))
y_full_pred_orig = scaler.inverse_transform(y_full_pred)

# Calculate metrics
full_rmse = np.sqrt(mean_squared_error(y_full_orig, y_full_pred_orig))
full_mae = mean_absolute_error(y_full_orig, y_full_pred_orig)
full_r2 = r2_score(y_full_orig, y_full_pred_orig)

print(f"\nFull Dataset Performance Metrics:")
print(f"RMSE: {full_rmse:.2f}")
print(f"MAE: {full_mae:.2f}")
print(f"R²: {full_r2:.4f}")

# Display a comparison of actual vs predicted values for the full training dataset
full_dates = train_df['date'].iloc[132:132+len(y)]

plt.figure(figsize=(14, 7))
plt.plot(full_dates, y_full_orig, label='Actual SSN')
plt.plot(full_dates, y_full_pred_orig, label='Predicted SSN')
plt.title('Full Dataset: Actual vs Predicted Sunspot Numbers')
plt.xlabel('Date')
plt.ylabel('Sunspot Number')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('full_dataset_performance.png')
plt.show()
```
